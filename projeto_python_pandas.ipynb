```python
#Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
plt.style.use("ggplot")
```

```python
dataframe = pd.read_excel(r"C:\Users\dpoli\OneDrive\Documentos\DIO\Projeto_analise_exploratoria_python_pandas\datasets\AdventureWorks.xlsx")
```

```python
dataframe.head(5)
```

```python
#Criando coluna de custo de cada entrada de venda
dataframe["Custo"] = dataframe["Custo Unitário"].mul(dataframe["Quantidade"])
```

```python
#Criando coluna de lucro para cada entrada de venda
dataframe["Lucro"] = dataframe["Valor Venda"] - dataframe["Custo"]
```

```python
#Criando coluna de tempo de envio para cada entrada de venda
dataframe["Tempo Envio"] = dataframe["Data Envio"] - dataframe["Data Venda"]
#Convertendo os dados da coluna "Tempo Envio" para integer
dataframe["Tempo Envio"] = (dataframe["Data Envio"] - dataframe["Data Venda"]).dt.days
dataframe.dtypes
```

```python
dataframe.isnull().sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

## Análise de Lucro

```python
tabela_lucro_marca_ano = dataframe.groupby([dataframe["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
tabela_lucro_marca_ano.to_csv("tabela_lucro_por_ano_marca.csv", index=False)
```

```python
dataframe.groupby([dataframe["Data Venda"].dt.year, "Marca"])["Lucro"].sum().plot.bar(title = "Gráfico de Lucro por Marca por Ano")
plt.xlabel("Marcas/Ano")
plt.ylabel("Lucro em milhões (R$)");
plt.savefig("gráfico_lucro_marca_ano.jpg")
```

```python
dataframe.groupby("Fabricante")["Lucro"].sum().plot.bar(title="Lucro por Marca Durante o Período dos Dados")
plt.xlabel("Marca")
plt.ylabel("Lucro em milhões (R$)")
plt.xticks(rotation="horizontal");
plt.savefig("gráfico_lucro_total_marca_2008_2009.jpg")
```

##### É possível notar que durante o período em que os dados foram coletados (2008 - 2009) a fabricante Fabrikam foi a que gerou maiores lucros nas vendas, mesmo tendo uma queda de um ano para o outro no lucro total anual.

## Análise de Produtos

```python
dataframe.groupby(dataframe["Produto"])["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Total Produtos Vendidos")
plt.xlabel("Unidades Vendidas")
plt.ylabel("Produto");
plt.savefig("gráfico_quantidade_total_produto.jpg")
```

```python
dataframe.groupby(dataframe["Produto"])["Lucro"].sum().sort_values(ascending=True).plot.barh(title = "Total Lucro por Produto")
plt.xlabel("Lucro Total por Produto")
plt.ylabel("Produto");
plt.savefig("lucro_total_por_produto.jpg")
```

#### É possível observar pelos dados que os produtos vendidos em maior quantidade não são necessáriamente os que deram maior lucro durante o período em que os dados foram coletados.

## Análise de Tempo de Envio por Marca

```python
estatisticas_tempo_envio = dataframe["Tempo Envio"].describe()
estatisticas_tempo_envio.to_csv("dados_tempo_envio.csv")
```

```python
dataframe.groupby("Fabricante")["Tempo Envio"].mean().plot.bar()
plt.title("Média de Tempo de Envio por Marca")
plt.xlabel("Marca")
plt.ylabel("Média de Tempo de Envio em Dias")
plt.xticks(rotation="horizontal");
plt.savefig("gráfico_média_tempo_envio.jpg")
```

```python
dataframe.groupby("Tempo Envio")["Tempo Envio"].count().plot.bar()
plt.title("Gráfico de Número de pedidos por Tempo de Envio")
plt.xlabel("Dias para Envio do Pedido")
plt.xticks(rotation="horizontal")
plt.ylabel("Número de Pedidos");
```

```python
dataframe[dataframe["Tempo Envio"] == 20]
```

#### Como é possível observar pelos dados coletados, a média de tempo de envio para cada uma das marcas é bem próximo, não apresentando diferença significativa de uma marca sobre outra. Além disso, é possível observar que o tempo de envio tem uma variação de 4 a 20 dias, porém, existe apenas uma entrada em que o envio demorou 20 dias, sendo que o próximo valor superior subsequente seria 13 dias, apontando que esse valor é um outlier, possivelmente um caso isolado de algum problema ocorrido durante o processamento da venda.
